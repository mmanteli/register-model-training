```python
import math

def dist(register):
    path=f'/scratch/project_462000353/amanda/register-training/register-model-training/sampling/distribution_results/eng_Latn_{register}.tsv'
    bins = []
    max_value = 0

    with open(path, 'r') as f:
        for line in f:
            num = int(line.split("\t")[1])
            if num>max_value: max_value=num
            if num==0: num+=1
            bin_to_put = int(math.log10(num))
            if len(bins) < bin_to_put + 1:
                for i in range(len(bins), bin_to_put+1):
                    bins.append(0)
            bins[bin_to_put] += 1

    values = [10**(i+1) for i in range(len(bins))]
    v_0 = 0
    print(f"Register: {register}")
    for v,b in zip(values, bins):
        print(f'{v_0}-{v}: \t {b}')
        v_0 = v
    print(f'maximum length: {max_value}')


for register in ["dtp", "HI", "HI-IN","IN","IP","LY", "NA", "ne", "no-label", "OP", "SP"]:
    dist(register)
    print("------------------------------")
```
